# 몬테카를로 방법으로 원주율 근사값 구하기

![대체 텍스트](https://wikidocs.net/images/page/21428/montecarlo.png)

$$ {원의 면적 \over  정사각형의 면적} = {원안의 점 개수 \over 정사각형안의 점 개수} $$


$$ {\pi r^2 \over  a^2 } = {원안의 점 개수 \over 정사각형안의 점 개수} $$

정사각형 한변의 크기 : 1,  원의 반지름 : 1/2

$$ {\pi (1/2)^2 \over  1^2 } = {원안의 점 개수 \over 정사각형안의 점 개수} $$

$$ {\pi  } = 4 *  {원안의 점 개수 \over 정사각형안의 점 개수} $$



첫번째, 필요한 라이브러리를 불러옴.

```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

정사각형의 크기, 원의 정사각형 안에 있는 점의 개수를 초기화

임의의 점의 개수를 나타내는 샘플 크기를 초기화

원의 사분면의 호를 정의

```python
square_size = 1
points_inside_circle = 0
points_inside_square = 0
sample_size = 100
arc = np.linspace(0, np.pi/2, 100)   # 원의 사분면의 호를 정의
```

사각형 내부에 임의의 점을 생성하는  generate_points() 함수 정의

```python
def generate_points(size):
    x = random.random()*size
    y = random.random()*size
    return (x, y)
```

생성된 점이 원 안에 속하는지를 확인하는 is_in_circle()함수

$$ x^2 + y^2  <=  r  $$

```python
def is_in_circle(point, size):
    return math.sqrt(point[0]**2 + point[1]**2) <= size
```

pi 값 을 계산하는 함수

```python
def compute_pi(points_inside_circle, points_inside_square):
    return 4 * (points_inside_circle / points_inside_square) 
```

샘플 크기 만큼 사각형 내부에 임의의 점을 생성하고 points_inside_square 변수를 증가시킴.

생성된 점이 원 안에 있다면 points_inside_circle변수를 증가

```python
plt.axes().set_aspect('equal')
plt.plot(1*np.cos(arc), 1*np.sin(arc))

sample_size = 10000

for i in range(sample_size):
     point = generate_points(square_size)
     plt.plot(point[0], point[1], 'c.')
     points_inside_square += 1
        
     if is_in_circle(point, square_size):
         points_inside_circle += 1

print("Approximate value of pi is {}"
.format(compute_pi(points_inside_circle, points_inside_square)))
```
